# Example 1 &ndash; Accuracy of the second-order model #

In this example, we reproduce figure 3 of the paper [*The determining role of covariances in large networks of stochastic neurons*](https://arxiv.org/abs/2212.09705). For details on PopNet's functions, see the [documentation](https://vincentpainchaud.github.io/PopNet/) of the package.

```python
import popnet as pn
import matplotlib.pyplot as plt
plt.ioff()
```

We start by defining a configuration with all parameters used in this example. For the code to run faster, we use a smaller network of 100 neurons than in the paper and we will run fewer stochastic simulations (only 100).

```python
config = pn.default_config('1I', scale='micro')
config.resize_network(100)
pop = config.network.populations[0]

pop.set_means(alpha=1.4, beta=2.5, gamma=1, theta=.75)
pop.set_scales(theta=.1)
config.network.c = 5.5
config.network.reset_parameters()

config.initial_state = .16, .51
config.set_covariances_from_expectations()
config.final_time = 20
config.delta = .02
config.executions = 100
```

Now, we integrate numerically the second-order system with PopNet's methods.

```python
integrator = pn.get_integrator(config, 'extended')
integrator.run('ode')
solution = integrator.output()
integrator.close()
```

We also run stochastic simulations and compute statistics.

```python
simulator = pn.get_simulator(config, act='step', mode='chain')
simulator.run(initial_state='random', verbose=True)
stats = simulator.output()
simulator.close()
```

Then, we define a Matplotlib figure on which we will draw the results.

```python
def get_handles_labels_with_column_titles(ax, titles):
    ncol = len(titles)
    empty_plots = ncol * [ax.plot([], [], marker='', ls='')[0]]
    handles, labels = ax.get_legend_handles_labels()
    col_len = len(handles) // ncol
    new_handles = []
    for j in range(ncol):
        new_handles += [empty_plots[j]] + handles[j*col_len : (j+1)*col_len]
    new_labels = []
    for j in range(ncol):
        new_labels += [titles[j]] + labels[j*col_len : (j+1)*col_len]
    return new_handles, new_labels

def remove_column_title_handles(legend):
    for vpack in legend._legend_handle_box.get_children():
        drawing_area = vpack.get_children()[0].get_children()[0]
        drawing_area.set_width(-4)
        
preamble = ('\\usepackage{lmodern}'
            '\\usepackage{amssymb}'
            '\\usepackage{sansmath}'
            '\\sansmath')
cm = 1 / 2.54 #in/cm

fs = 8
hpad = .4
hlen = 1.4

fig, axes = pn.figure(subplots=((2,3,(1,2)), (2,3,(4,5)), (2,3,(3,6))), figsize=(18*cm, 5*cm),
                      font_family='sans-serif', usetex=True, preamble=preamble, tight_layout=False)
fig.subplots_adjust(left=.037, right=.99, bottom=.15, top=.98, wspace=.25, hspace=.1)

solution.fig = stats.fig = fig
```

On the first subfigure, we plot the expectations of the fractions of the network in each of the three states that we obtained from the dynamical system's solution as well as the corresponding averages obtained from stochastic simulations.

```python
stats.ax = axes[0]
solution.ax = axes[0]

solution.plot_expectations(ls='dotted', zorder=2.5)
for X in ('A', 'R', 'S'):
    lighter_color = stats.colors[f'C{X}{X}'][0][0]
    stats.fill[X](color=lighter_color, alpha=.2)
    stats.plot[X](color=lighter_color)
solution.setup(set_xlabel=False, fontsize=fs)
solution.ax.set_xticklabels([])
solution.ax.tick_params(labelsize=fs)
solution.ax.set_xlim([0, 16])
solution.ax.set_ylim([-.05, 1.05])
titles = ['Solution', 'Statistics']
handles, labels = get_handles_labels_with_column_titles(solution.ax, titles)
legend = solution.ax.legend(handles, labels, fontsize=fs, ncol=2, handletextpad=hpad,
                            handlelength=hlen, loc='center right')
remove_column_title_handles(legend)
```

On the second subfigure, we plot the variances from the solution and the sample variances from the stochastic simulations.

```python
stats.ax = axes[1]
solution.ax = axes[1]

for X in ('A', 'R', 'S'):
    solution.plot[f'C{X}{X}'](ls='dotted', color=solution.colors[X][0], zorder=2.5)
stats.plot_variances()
solution.setup(units='$1/\\gamma$', ylim='unbounded', fontsize=fs, labelpad=2)
solution.ax.tick_params(labelsize=fs)
solution.ax.set_xlim([0, 16])
titles = ['Solution', 'Statistics']
handles, labels = get_handles_labels_with_column_titles(solution.ax, titles)
legend = solution.ax.legend(handles, labels, fontsize=fs, ncol=2, handletextpad=hpad,
                            handlelength=hlen, loc='upper right')
remove_column_title_handles(legend)
```

On the third subfigure, we plot a phase plane of the mean-field system, on which we draw its solution as well as a single trajectory of the Markov chain.

```python
plane = pn.get_system(config, 'mean-field').get_phase_plane((0,1))

plane.fig = fig
plane.ax = axes[-1]
xlim = (-.01, .3)
ylim = (-.02, .55)

plane.streamplot((100, 100), xlim=xlim, ylim=ylim, color='black', linewidth=.6, arrowsize=.5, density=1.25)
plane.plot_nullclines(lw=1.5)
plane.plot_solution(lw=1.5, label='Mean-field')
plane.plot_trajectory(lw=.8)
plane.setup(xlim=xlim, ylim=ylim, fontsize=fs, xlabel_kwargs={'labelpad': 2}, ylabel_kwargs={'va': 'top'})
plane.legend(lw=1.5, fontsize=fs, loc='lower right', handletextpad=hpad, handlelength=hlen)
plane.ax.tick_params(labelsize=fs)
```

Finally, we draw the figure. Of course, since the network is smaller, we performed less simulations, and the results are random, the figure will be slightly different from that shown in the paper.

```python
pn.draw()
```
